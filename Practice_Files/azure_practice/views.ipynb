## CREATE VIEW

```python
%sql
CREATE OR REPLACE VIEW low_paid_employees
    (id COMMENT 'Unique identification number', firstName)
    COMMENT 'View for experienced employees'
    AS SELECT id, firstName
         FROM people_10m
        WHERE salary < 50000;
```

```python
%sql
SELECT * FROM low_paid_employees;
```

## TEMPORARY VIEW

```python
%sql
CREATE TEMPORARY VIEW high_salary_employees AS
SELECT id, firstName, lastName, salary
FROM people_10m
WHERE salary > 50000;
```

```python
%sql
SELECT * FROM high_salary_employees;
```
